```python
fund_master["launch_date"] = pd.to_datetime(fund_master["launch_date"])

fund_master.to_csv("01_fund_master_cleaned.csv", index=False)
```

```python
nav_history["date"] = pd.to_datetime(nav_history["date"])

nav_history = nav_history.sort_values(["amfi_code","date"])

nav_history["nav"] = nav_history.groupby("amfi_code")["nav"].ffill()

nav_history = nav_history[nav_history["nav"] > 0]

nav_history.to_csv("02_nav_history_cleaned.csv", index=False)
```

```python
aum_by_fund_house["date"] = pd.to_datetime(aum_by_fund_house["date"])

aum_by_fund_house = aum_by_fund_house[
    aum_by_fund_house["aum_lakh_crore"] > 0
]

aum_by_fund_house.to_csv("03_aum_by_fund_house_cleaned.csv", index=False)
```

```python
monthly_sip_inflows["month"] = pd.to_datetime(
    monthly_sip_inflows["month"]
)

monthly_sip_inflows.to_csv(
    "04_monthly_sip_inflows_cleaned.csv",
    index=False
)
```

```python
category_inflows["month"] = pd.to_datetime(
    category_inflows["month"]
)

category_inflows.to_csv(
    "05_category_inflows_cleaned.csv",
    index=False
)
```

```python
industry_folio_count["month"] = pd.to_datetime(
    industry_folio_count["month"]
)

industry_folio_count.to_csv(
    "06_industry_folio_count_cleaned.csv",
    index=False
)
```

```python
numeric_cols = [
    "return_1yr_pct",
    "return_3yr_pct",
    "return_5yr_pct",
    "benchmark_3yr_pct",
    "alpha",
    "beta",
    "sharpe_ratio",
    "sortino_ratio",
    "std_dev_ann_pct",
    "max_drawdown_pct",
    "aum_crore",
    "expense_ratio_pct",
    "morningstar_rating"
]

scheme_performance[numeric_cols] = scheme_performance[numeric_cols].apply(pd.to_numeric)

scheme_performance = scheme_performance[
    (scheme_performance["expense_ratio_pct"] >= 0.1) &
    (scheme_performance["expense_ratio_pct"] <= 2.5)
]

scheme_performance.to_csv(
    "07_scheme_performance_cleaned.csv",
    index=False
)
```

```python
investor_transactions["transaction_date"] = pd.to_datetime(
    investor_transactions["transaction_date"]
)

investor_transactions["transaction_type"] = (
    investor_transactions["transaction_type"]
    .str.strip()
    .str.title()
)

investor_transactions = investor_transactions[
    investor_transactions["amount_inr"] > 0
]

investor_transactions.to_csv(
    "08_investor_transactions_cleaned.csv",
    index=False
)
```

```python
portfolio_holdings["portfolio_date"] = pd.to_datetime(
    portfolio_holdings["portfolio_date"]
)

portfolio_holdings = portfolio_holdings[
    portfolio_holdings["weight_pct"] > 0
]

portfolio_holdings.to_csv(
    "09_portfolio_holdings_cleaned.csv",
    index=False
)
```

```python
benchmark_indices["date"] = pd.to_datetime(
    benchmark_indices["date"]
)

benchmark_indices = benchmark_indices.sort_values(
    ["index_name", "date"]
)

benchmark_indices = benchmark_indices[
    benchmark_indices["close_value"] > 0
]

benchmark_indices.to_csv(
    "10_benchmark_indices_cleaned.csv",
    index=False
)
```

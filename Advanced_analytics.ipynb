```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python


investor = pd.read_csv("08_investor_transactions_cleaned.csv")

performance = pd.read_csv("07_scheme_performance_cleaned.csv")

portfolio = pd.read_csv("09_portfolio_holdings_cleaned.csv")
```

```python
investor["transaction_type"].value_counts()
```

```python
investor["transaction_type"].value_counts().plot(
    kind="bar",
    figsize=(6,4),
    title="Transaction Type Distribution"
)

plt.show()
```

```python
state = investor.groupby("state")["amount_inr"].sum()

state.sort_values(ascending=False)
```

```python
state.sort_values().plot(
    kind="barh",
    figsize=(8,5),
    title="Investment by State"
)

plt.show()
```

```python
age = investor.groupby("age_group")["amount_inr"].mean()

age
```

```python
age.plot(
    kind="bar",
    figsize=(6,4),
    title="Average Investment by Age Group"
)

plt.show()
```

```python
investor["payment_mode"].value_counts()
```

```python
investor["payment_mode"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)

plt.show()
```

```python
sector = portfolio.groupby("sector")["weight_pct"].mean()

sector.sort_values(ascending=False)
```

```python
sector.sort_values().plot(
    kind="barh",
    figsize=(8,5),
    title="Average Sector Weight"
)

plt.show()
```

```python
top_stock = portfolio.sort_values(
    by="market_value_cr",
    ascending=False
).head(10)

top_stock
```

```python
top_stock.plot(
    x="stock_name",
    y="market_value_cr",
    kind="bar",
    figsize=(8,5)
)

plt.show()
```

```python
performance["risk_grade"].value_counts()
```

```python
performance["risk_grade"].value_counts().plot(
    kind="bar",
    figsize=(6,4)
)

plt.show()
```
